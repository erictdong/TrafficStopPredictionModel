```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix, f1_score, classification_report, make_scorer
from sklearn import model_selection
```

```python
df = pd.read_csv('finalcsv.csv')
df.head()
```

```python
# define inputs and target
inputs = df[df.columns[1:18]]
target = df['citation']
```

```python
# test train split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.1)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
## KNN model, k=5
k = 5
neigh = KNeighborsClassifier(n_neighbors = k)
neigh.fit(X_train, y_train)
y_hat = neigh.predict(X_test)
```

```python
# model evaluation
score = neigh.score(X_test, y_test)
print('The mean accuracy score on this model is:' , score)
print('Confusion matrix:')
print(confusion_matrix(y_test, y_hat))
target_names = ['Citation given', 'Citation not given']
print('Classification report:')
print(classification_report(y_test, y_hat, target_names=target_names))
```

```python
## Naive bayes model
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_gnb = gnb.predict(X_test)
```

```python
# model evaluation
gnbscore = gnb.score(X_test, y_test)
print('The mean accuracy score on this model is:' , gnbscore)
print('Confusion matrix:')
print(confusion_matrix(y_test, y_gnb))
target_names = ['Citation given', 'Citation not given']
print('Classification report:')
print(classification_report(y_test, y_gnb, target_names=target_names))
```

```python
## Logistic regression model
lr = LogisticRegression(solver = 'liblinear')
lr.fit(X_train, y_train)
y_lr = lr.predict(X_test)
```

```python
# model evaluation
lrscore = lr.score(X_test, y_test)
print('The mean accuracy score on this model is:', lrscore)
print('Confusion matrix:')
print(confusion_matrix(y_test, y_lr))
target_names = ['Citation given', 'Citation not given']
print('Classification report:')
print(classification_report(y_test, y_lr, target_names=target_names))
```
